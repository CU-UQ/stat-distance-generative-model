## Generative Models
- **Explicit models:** Compute density function $p(x)$ exactly, e.g. directed probabilistic graphical models, autoregressive model, normalizing flow;
- **Approximate models:** Maximize an MLE approximation or lower bound, e.g. VAE, energy-based model, diffusion model;
- **Implicit models:** No well-defined density function, e.g. GAN.

## Kernel density estimation
Compare density functions directly?

If the density function is NOT explicit but sampling is cheap, we can estimate an approximated density function using kernel density estimation (KDE), which has the form
$$\hat{p}(x) = \frac{1}{Nh_N}\sum_{n=1}^N K(\frac{x - X_n}{h_N}),$$
where $h$ is the bandwidth, $K$ is the kernel function such that $\int K(x) dx = 1$ and $\int xK(x)dx=0$.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from sklearn.neighbors import KernelDensity

# Generate true pdf and evaluation points
x = np.linspace(-5,10,1000)
y = 0.2 * norm(1, 1).pdf(x) + 0.8 * norm(5, 1).pdf(x)

# Sampling
N = 100 # number of samples
X = np.concatenate((np.random.normal(1,1,int(0.2*N)), np.random.normal(5,1,int(0.8*N))))

# Gaussian KDE
kde = KernelDensity(kernel="gaussian", bandwidth=N**(-0.2)).fit(X[:,None])
log_dens = kde.score_samples(x[:,None])

# Plot
fig, ax = plt.subplots(figsize=(12,8))
ax.fill(x, y, fc="black", alpha=0.2, label="input distribution")
ax.plot(x,np.exp(log_dens),label='Gaussian KDE')
ax.legend(loc="upper left")
ax.plot(X, -0.005 - 0.01 * np.random.random(X.shape), "+k")
plt.show()
```

The estimator $\hat{p}(x)$ is biased but asymtotically unbiased. 

**Theorem (curse of dimensionality):** With some regularity conditions on kernel function $K$, bandwidth $h_N\to 0$, and $N\cdot h_N^D\to\infty$ with data dimension $D$, we have
$$\sqrt{Nh_N^D}\big(\hat{p}(x) - \mathbb{E}[\hat{p}(x)]\big)\xrightarrow[]{d}\mathcal{N}\bigg(0,p(x)\int K^2(y)dy\bigg),$$
where $x$ is a point of continuity of $f$. 
Because KDE is aysymtotically unbiased, the theorem above shows that the consistency of the estimator $\hat{f}(x)$ (the estimator converges to true value $f(x)$ in probability).

## Statistical distance
In mathematical statistics, statistical distance is a quantity representing the difference between two random variables, or two distributions, or two groups of empirical samples. 
Statistical distance is often used for loss function or evaluation criterion for generative models.
There are mainly two categories of statistical distance. 

### f-divergence
Assuming density functions exist, an **f-divergence** is
\begin{equation}
    D_f(p\|q) = \int_\Omega f\bigg(\frac{q(x)}{p(x)}\bigg)p(x) dx,
\end{equation}
where $\Omega$ is the domain (usually a subset of $\mathbb{R}^D$), $f$ is convex and $f(1)=0$.
The non-negativity can be obtained by Jensen's inequality
$$D_f(p\|q) = \int_\Omega f\bigg(\frac{q(x)}{p(x)}\bigg)p(x) dx \geq f\bigg(\int_\Omega \frac{q(x)}{p(x)}p(x) dx\bigg) = f(1) = 0.$$
The most well-known f-divergence is the **Kullback-Leibler divergence**, also known as KL divergence, relative entropy. 
It chooses function $f = -\log$. In fact, maximum likelihood estimation (MLE) can be interpretated as a KL divergence minimization approach. 

Given an unknown oracle density $p(x)$, and our surrogate density $q(x)$, minimizing their KL divergence (with the specific order) is 
\begin{align}
    \text{KL}(p\|q) &= \int_\Omega \log\bigg(\frac{p(x)}{q(x)}\bigg)p(x) dx\\
                             &= \int_\Omega \log \big(p(x)\big) p(x) dx - \mathbb{E}_{p} \big [\log \big(q(x)\big)\big]\\
                             &= - \mathbb{E}_{p} \big [\log \big(q(x)\big)\big] + C\\
                             &\approx -\frac{1}{m} \sum_{i=1}^m \log \big(q(X_i)\big) + C,
\end{align}
where $\{X_i\}_{i=1}^m$ are samples from $p(x)$. 
As a result, finding $q^* = \arg\max_q \log \big(q(x)\big)$ using samples from $p(x)$ is equivalent with solving $q^* = \arg\min_q \text{KL}(p\|q)$. 
KL divergence is **asymmetric**. 
If we flip the order of inputs, we have another estimator
\begin{align}
    \text{KL}(q\|p) &= \int_\Omega \log\bigg(\frac{q(x)}{p(x)}\bigg)q(x) dx\\
                             &= -\mathbb{H} \big[q(x)\big] - \mathbb{E}_{q} \big [\log \big(p(x)\big)\big]\\
                             &\approx \frac{1}{n} \sum_{i=1}^n \log\bigg(\frac{q(Y_i)}{p(Y_i)}\bigg),
\end{align}
where $\{Y_i\}_{i=1}^n$ are samples from $q(x)$. 
The difference between them is that minimizing $\text{KL}(p\|q)$ requires samples from $p(x)$ but not closed form (or approximated) density function, while minimizing $\text{KL}(q\|p)$ needs density function but not samples. 
The estimators are consistent by law of large number, but estimating $\text{KL}(q\|p)$ consistently is only based on accurate density function $p(x)$, which is not feasible in practice. 

```python
from scipy import stats
from scipy.optimize import minimize

## A Gaussian parameteric estimation to minimize KL(p\|q)
## As we show, it is equivalent with MLE
def gaussian_pvq(params):
    mean = params[0]   
    sd = params[1]

    # Calculate negative log likelihood
    nll = -np.sum(stats.norm.logpdf(X, loc=mean, scale=sd))

    return nll

initParams = [1, 1]
pvq_results = minimize(gaussian_pvq, initParams, method='Nelder-Mead')
pvq_y = norm(pvq_results.x[0],pvq_results.x[1]).pdf(x)

## A Gaussian parametric estimation to minimize KL(q\|p)
## It has two local minimum
def gaussian_qvp(params):
    mean = params[0]   
    sd = params[1]
    
    # Sample from q
    Y = np.random.normal(mean,sd,100)

    # Calculate negative log likelihood
    kl_qvp = -np.sum(stats.norm.logpdf(Y, loc=np.mean(X), scale=np.std(X)) 
                     - stats.norm.logpdf(Y, loc=np.mean(Y), scale=np.std(Y)))

    return kl_qvp

# 1st local minimum
initParams1 = [8, 1]
qvp_results1 = minimize(gaussian_qvp, initParams1, method='Nelder-Mead')
qvp_y1 = norm(qvp_results1.x[0],qvp_results1.x[1]).pdf(x)

# 2nd local minimum
initParams2 = [1, 1]
qvp_results2 = minimize(gaussian_qvp, initParams2, method='Nelder-Mead')
qvp_y2 = norm(qvp_results2.x[0],qvp_results2.x[1]).pdf(x)

## Plot two results
fig, ax = plt.subplots(figsize=(12,8))
ax.fill(x, y, fc="black", alpha=0.2, label="true $p(x)$")
ax.plot(x,pvq_y,label=r'$\arg\min_q$ KL($p\|q$)',c='b')
ax.plot(x,qvp_y1,label=r'$\arg\min_q$ KL($q\|p$)_1',c='r')
ax.plot(x,qvp_y2,label=r'$\arg\min_q$ KL($q\|p$)_2',c='g')
ax.legend()
plt.show()
```

To deal with problems from assymetry, **Jensen-Shannon divergence** is purposed as follow
\begin{equation}
    \text{JS}(p\|q) = \text{KL}(p\|\frac{p+q}{2}) + \text{KL}(q\|\frac{p+q}{2}),
\end{equation}
where $\frac{p+q}{2}$ represents the mixture density function
$$\big(p+q\big) (x) = \frac{1}{2}\big(p(x)+q(x)\big).$$
Jensen-Shannon divergence is used for virtual training criterion for (original) GAN \[Theorem 1, Goodfellow et al. 2014\]. 
However, computing $p+q$ requires closed form for both $p(x)$ and $q(x)$. 

There are many other examples of f-divergence:
- Hellinger distance
\begin{equation}
    H^2(p,q) = \int_\Omega \bigg(\sqrt{\frac{p(x)}{q(x)})} - 1\bigg)^2 q(x) dx
\end{equation}
- Total variation
\begin{equation}
    \text{TV}(p,q) = \int_\Omega \bigg|\frac{p(x)}{q(x)}-1\bigg| q(x) dx
\end{equation}

One disadvantage of f-divergence is that we need an explicit or approximated density function. We also have to assume two inputs $p(x)$ and $q(x)$ sharing the same support (or say $p\ll q$ from **Radon-Nikodym theorem**, replace $\frac{p(x)}{q(x)}$ with Radon-Nikodym derivative $\frac{d\mu_p}{d\mu_q}$).

### Integral probability metrics 
In general, integral probability metrics (IPM) have the following form
\begin{equation}
    \text{IPM}_{\mathcal{F}} (p,q) = \sup_{f\in\mathcal{F}} \mathbb{E}_{p(x)}[f(x)] - \mathbb{E}_{q(y)}[f(y)],
\end{equation}
where $\mathcal{F}$ is a function class that controls the performance of its IPM. 
A commonly used IPM is **Wasserstein distance**, which is also known as earth mover distance, or Kantorovich distance. Wasserstein distance is defined as 
\begin{equation}
    W_1(p,q) = \inf_{\pi(x,y)\in\Pi\big(p(x),q(y)\big)} \int_{\Omega\times\Omega} \rho(x - y) d\pi(x,y),
\end{equation}
where $\pi$ is a joint distribution with marginals being $p(x)$ and $q(y)$, $\rho:\Omega\times\Omega\to\mathbb{R}$ is a metric on $\Omega$. 
By Kantorovich-Rubenstein duality theorem, $W_1$ is an IPM with $\mathcal{F}=\{f: f\text{ is Lipschitz continuous with Lipschitz constant bounded by }1\}$.
The intuition behind Wasserstein distance is shown in the following figure.

![Wasserstein distance figure](wasserstein.png)
Figure from [Wikipedia page](https://en.wikipedia.org/wiki/Wasserstein_metric)

Wasserstein distance has many nice properties, e.g. it implies weak convergence of random variables. 
However, computation of Wasserstein distance for high dimension is not feasible. 
We can only do 1D Wasserstein distance directly. 
1D Wasserstein distance can be solved using Sinkhorn algorithm or Hungarian algorithm iteratively, and their results are biased.

Another computationally feasible IPM is **maximum mean discrepancy** (MMD), also known as kernel embedded distance, kernel Stein discrepancy. 
Different from Wasserstein distance, we choose function class $\mathcal{F} = \mathcal{H}_k$, where $\mathcal{H}_k$ is the reproducing kernel Hilbert space (RKHS) generated from kernel $k$, then we get
\begin{equation}
    \text{MMD}(p,q) = \sup_{\lVert f\rVert_{\mathcal{H}_k}=1} \mathbb{E}_p[f(x)] - \mathbb{E}_q[f(y)] = \lVert \Psi(p) - \Psi(q)\rVert_\mathcal{H},
\end{equation}
where $\Psi:\mathcal{P}(\Omega)\to\mathcal{H}_k$ defined as $\Psi(p) = \int_\Omega k(x,\cdot)p(x)dx$. 
The advantage of MMD is its computational convenience. 
With samples $\{X_i\}_{i=1}^m\sim p(x)$ and $\{Y_j\}_{j=1}^n\sim q(y)$, an unbiased estimator of MMD$^2$ is 
\begin{align}
    \text{MMD}^2(p,q) &= \lVert \Psi(p) - \Psi(q)\rVert^2_\mathcal{H}\\
                    &\approx \frac{1}{m(m-1)}\sum_{i,j=1,i\neq j}^m k(X_i,X_j) - \frac{2}{mn}\sum_{i=1}^m\sum_{j=1}^n k(X_i,Y_j) +\frac{1}{n(n-1)}\sum_{i,j=1,i\neq j}^n k(Y_i,Y_j).
\end{align}

The statistical properties of the MMD estimator [Theorem 10 and Corollary 16, Gretton et al. 2012] are
\begin{align}
    \mathbb{P}\big[\vert\widehat{\text{MMD}^2} - \text{MMD}^2\vert > t\big]&\leq 2\exp\bigg(-\frac{t^2m}{16K^2}\bigg)\\
    m^{1/2}\big(\widehat{\text{MMD}^2} - \text{MMD}^2\big)&\xrightarrow[]{d}\mathcal{N}(0,\sigma_U),
\end{align}
where $\widehat{\text{MMD}}^2$ is the MMD estimator, $K$ is the maximal value of kernel function, $m$ is the number of samples (assuming $m=n$), and $\sigma_U$ is some bounded quantity.
The result shows the estimator is consistent, and its convergence rate is independent with dimension $D$. 
The disadvantages of MMD is that its performance highly depends on kernel and bandwidth choice. 

```python
from scipy.stats import wasserstein_distance 
from MMD_func import MMD 

### Conduct two-sample-test for Wasserstein distance, MMD
N = 100 # number of samples
T = 100 # number of trials

## Experiment 1: standard Gaussian and standard Gaussian
w_1,mmd1_1,mmd2_1 = np.empty(shape=T),np.empty(shape=T),np.empty(shape=T)
for t in range(T):
    X1 = np.random.normal(0,1,int(N))
    X2 = np.random.normal(0,1,int(N))
    w_1[t] = wasserstein_distance(X1,X2)
    mmd1_1[t] = MMD(X1[:,None],X2[:,None],[0.01])
    mmd2_1[t] = MMD(X1[:,None],X2[:,None],[10.0])
    
## Experiment 2: Gaussian and Gaussian mixtures
w_2,mmd1_2,mmd2_2 = np.empty(shape=T),np.empty(shape=T),np.empty(shape=T)
for t in range(T):
    X1 = np.random.normal(pvq_results.x[0],pvq_results.x[1],int(N))
    X2 = np.concatenate((np.random.normal(1,1,int(0.2*N)), np.random.normal(5,1,int(0.8*N))))
    w_2[t] = wasserstein_distance(X1,X2)
    mmd1_2[t] = MMD(X1[:,None],X2[:,None],[0.01])
    mmd2_2[t] = MMD(X1[:,None],X2[:,None],[10.0])

## Plot two results
fig, axs = plt.subplots(1,3,figsize=(15,4))
axs[0].boxplot([w_1,w_2])
axs[0].set_xticklabels([r'NvN',r'NvGM'])
axs[0].set_title(r'Wasserstein distance')
axs[1].boxplot([mmd1_1,mmd1_2])
axs[1].set_xticklabels([r'NvN',r'NvGM'])
axs[1].set_title(r'MMD, bandwidth=0.01')
axs[2].boxplot([mmd2_1,mmd2_2])
axs[2].set_xticklabels([r'NvN',r'NvGM'])
axs[2].set_title(r'MMD, bandwidth=10.0')
plt.show()
```

## Evaluation methods in practice
### Inception score
For **supervised** generative models, inception score measures the label difference. It is defined as 
$$  \text{IS} = \exp\big[\mathbb{E}_{p_\theta(x)}\text{KL}\big(p(y\vert x)||p_\theta(y)\big)\big],$$
where $p_\theta(y)=\int p(y\vert x)p_\theta(x)dx$. 
Larger IS means better performance. 

### Frechet inception distance
Frechet inception distance (FID) is a parametric estimation of Wasserstein distance. 
Assuming two inputs are both Gaussian, W-2 Wasserstein distance is 
\begin{equation}
    W_2^2\big(\mathcal{N}(\mu_1,\Sigma_1),\mathcal{N}(\mu_2,\Sigma_2)\big) = \lVert \mu_1 - \mu_2\rVert^2_2 + \text{Tr}\big(\Sigma_1+\Sigma_2-2(\Sigma_1\Sigma_2)^{1/2}\big). 
\end{equation}
Some authors use pre-trained function $f$ (usually neural-network) and compute FID between $\{f(X_i)\}_{i=1}^m$ and $\{f(Y_j)\}_{j=1}^n$. 
The complexity of computing FID is $O(nd^2+d^3)$.

### Kernel inception distance
Kernel inception distance (KID) is MMD with specific kernel $k(x,y) = \bigg(\frac{1}{d}x^Ty + 1\bigg)^3$.
The complexity of computing KID is $O(n^2d)$.

### Heuristic ways
- Compression rate;
- Distanglement;
- Clustering in latent space (with labels only).
